```python
import random
import numpy as np
from matplotlib.pyplot import show
def dice_normaldist(number_of_dice, rolling):
    hasil = []
    dadu = [1,2,3,4,5,6]  
    result = 0
    for i in range(rolling):
        for j in range(number_of_dice):
            result += np.random.choice(dadu)
        hasil.append(result)
        result = 0
#     plt.figure(figsize=(10,8))
#     sns.distplot(hasil)
#     plt.xticks(range(number_of_dice, (number_of_dice*6)+1))
    return hasil
```

```python
%%time
data = dice_normaldist(4, 1000000)
hist, hist_edges = np.histogram(data, bins=10, normed = True)
dx = hist_edges[1] - hist_edges[0]
F1 = np.cumsum(hist)*dx

plt.plot(hist_edges[1:],F1)
```
